We would now want to collapse them to should just the counts for the breeds (we don't need to know what district has what count)

## Taranaki Aggregration

```r
taranaki_df <- aggregate(taranaki['Value'], by=taranaki['Category'], sum) 
taranaki_df
taranaki_pop_agg <- aggregate(taranaki_pop["Value"], by = taranaki_pop['Category'], sum)
taranaki_pop_agg
```

Rename the column variables to match what has been done before

```r
taranaki_df <- rename(taranaki_df, Breed = Category, Taranaki = Value) 
```

```r
taranaki_df
```

Repeat this for all of the regions
## Tasman

```r
tasman_df <- aggregate(tasman['Value'], by=tasman['Category'], sum) 
tasman_df <- rename(tasman_df, Breed = Category, Tasman = Value) 
tasman_df
```

## Waikato

```r
waikato_df <- aggregate(waikato['Value'], by=waikato['Category'], sum) 
waikato_df <- rename(waikato_df, Breed = Category, Waikato = Value) 
waikato_df
waikato_pop_agg <- aggregate(waikato_pop["Value"], by = waikato_pop['Category'], sum)
waikato_pop_agg
```

## Wellington

```r
wellington_df <- aggregate(wellington['Value'], by=wellington['Category'], sum) 
wellington_df <- rename(wellington_df, Breed = Category, Wellington = Value) 
wellington_df
wellington_pop_agg <- aggregate(wellington_pop["Value"], by = wellington_pop['Category'], sum)
wellington_pop_agg
```

## West Coast

```r
west_coast_df <- aggregate(west_coast['Value'], by=west_coast['Category'], sum) 
west_coast_df <- rename(west_coast_df, Breed = Category, West_Coast = Value) 
west_coast_df
west_coast_pop_agg <- aggregate(west_coast_pop["Value"], by = west_coast_pop['Category'], sum)
west_coast_pop_agg
```

## Finally lets combine them all into one dataframe

```r
(taranaki_df)
```

```r
tar_tas <- taranaki_df %>%
 left_join(tasman_df, by = "Breed")

tar_tas_wai<- tar_tas %>%
 left_join(waikato_df, by = "Breed")

tar_tas_wai_wel <- tar_tas_wai %>%
 left_join(wellington_df, by = "Breed")

tar_to_wc_pets <- tar_tas_wai_wel %>%
 left_join(west_coast_df, by = "Breed")
```
