## Next we would want to change the dog breeds to be on columns and the regions on the rows ## 

This will make it easier to clean out the breeds that we do not need and to make plotting easier

```r
#transpose the dataframe to get the breeds as columns and the regions as rows
t_pets_pc = t(pets_pc)
t_pets_pc = as.data.frame(t_pets_pc)
```

```r
t_pets_pc
```

```r
#Next we will load the janitor library to help wrangle this transposed dataframe into a more usable dataframe
library(janitor)
```

```r
#this will bring the 2nd row of the dataframe to become the column names of the dataframe
t_pets_pc = row_to_names(t_pets_pc,2)
t_pets_pc
```

```r
#Now we must rename the first column as currently it is nonsensical, it should be regions
t_pets_pc = t_pets_pc%>%
    rename( Region = Breed)
```

```r
t_pets_pc
```

```r
#currently all of the columns are factors, so we will load the hablar library to use the retype function
#which will retype them to the most likely type 
library(hablar)
t_pets_pc <- t_pets_pc %>%
                    retype()
```

```r
t_pets_pc
#this has changed them to doubles/ints and characters as they should be 
```

## Now to remove dogs who aren't present in any region

```r
#Remove the dogs who have a count of 0 in all of the regions 
t_pets_pc$`Brazilian Fila` = NULL
t_pets_pc$`Perro de Presa Canario` = NULL
t_pets_pc$`Tosa, Japanese` = NULL
```

```r
t_pets_pc
```

## And we will combine breeds into a bigger breed present in the AKC dataset where they have been split into more specific breed in thisdatseset 

```r
#We want to combine long coat and smooth coat chihuahua into one column called "Chihuahua".
#this does elementwise addition when making the new column
t_pets_pc = t_pets_pc %>%
                        mutate(Chihuahua = t_pets_pc$`Chihuahua, Long Coat` + t_pets_pc$`Chihuahua, Smooth Coat`)
```

```r
t_pets_pc
```

```r
#remove the original chihuahuas from the dataframe
t_pets_pc$`Chihuahua, Long Coat` = NULL
t_pets_pc$`Chihuahua, Smooth Coat` = NULL
```

```r
#combine the collies into one variable 
#This is doing the same we did for the chihuahuas but for the smooth and rough collies
t_pets_pc = t_pets_pc %>%
                        mutate(Collie = t_pets_pc$`Collie, Smooth` + t_pets_pc$`Collie, Rough`)
```

```r
t_pets_pc
```

```r
#And remove the smooth collie and the rough collie from the dataframe
t_pets_pc$`Collie, Smooth` = NULL
t_pets_pc$`Collie, Rough` = NULL
```

```r
t_pets_pc
```

### The Lab Retriever is the most common dog breed in all regions. In order to figure out the weight of the second most common dog, I removed the column 'Labrador Retriever' and sorted for max in all other rows

```r
doggums_df_2 = condensed_further_df
```

```r
doggums_df_2['Labrador Retriever'] <- NULL 
```

```r
#We want to know what the second most popular dog in each region is.
#This code goes through each region, and tells us the second most common dog breed in each region. 
for (i in 1:16){
    second_most_common_dog = doggums_df_2 %>%
      rowid_to_column() %>%
      filter(rowid == i) %>%
      gather(col, val, Beagle:Collie) %>%
      filter(abs(val) == max(abs(val))) %>%
      select(-rowid, -val)
    print(second_most_common_dog$col)
}
```
