## Pets Per Capita

Now we want to divide the counts of each breed of dog in a region by the population of each region. This will give a per capita count of each breed within a region (along with total counts of all breeds). This allows for comparison of breeds across regions

```r
#here we will mutate in a new variable that is region_per_capita for each region
#this takes a column (region) in the pets dataframe and divides each element in the column by the value of the 
#populations dataframe for the same region, giving a per capita count for each row in the column. 
pets_with_pc = pets %>%
        mutate(
            Auckland_Per_Capita =  pets$Auckland / populations$Auckland,
            Bay_Of_Plenty_Per_Capita = pets$Bay_of_Plenty / populations$Bay_Of_Plenty,
            Canterbury_Per_Capita = pets$Canterbury / populations$Canterbury,
            Gisbourne_Per_Capita = pets$Gisbourne / populations$Gisbourne,
            Hawkes_Bay_Per_Capita = pets$Hawkes_Bay / populations$Hawks_Bay,
            Taranaki_Per_Capita = pets$Taranaki / populations$Taranaki,
            Tasman_Per_Capita = pets$Tasman / populations$Tasman,
            Waikato_Per_Capita = pets$Waikato / populations$Waikato,
            Wellington_Per_Capita = pets$Wellington / populations$Wellington,
            West_Coast_Per_Capita = pets$West_Coast / populations$West_Coast,
            Manawatu_Whanganui_Per_Capita = pets$Manawatu_Whanganui / populations$Manawatu_Whanganui,
            Marlborough_Per_Capita = pets$Marlborough / populations$Marlborough,
            Nelson_Per_Capita = pets$Nelson / populations$Nelson,
            Otago_Per_Capita = pets$Otago / populations$Otago,
            Southland_Per_Capita = pets$Southland / populations$Southland,
            Northland_Per_Capita = pets$Northland / populations$Northland
            
    )
pets_with_pc
```

```r
#Now we will make a new dataframe that is the per capita dataframe, by only selecting the columns that had the
#per capita information from the previous dataframe.
pets_pc = pets_with_pc[,c(2,19:34)]
pets_pc
```

## We Have the pets per capita for each region!

## Now we just want the home ownership per capita for each region

We want to simplify the way it came in the csv to just owned(atleast partially) or rented

```r
#The owned homes are from the first and third rows, as these show homes that are partially or fully owned 
#Whilst the second row has informatino about homes that aren't owned in anyway
Owned = homes[1,2:17] + homes[3,2:17]
Rented = homes[2,2:17]
```

```r
#Then we will make a dataframe out of these two variables by binding them together 
ownership = rbind(Owned,Rented)
ownership
```

Next we want to divide each value by the population size, this is done by column wise division

```r
#here we will mutate in a new variable that is region_per_capita for each region
#this takes a column (region) in the ownership dataframe and divides each element in the column by the value of the 
#populations dataframe for the same region, giving a per capita count for each row in the column. 
ownership_with_pc = ownership %>%
            mutate(
                Auckland_Per_Capita = ownership$Auckland / populations$Auckland,
                Bay_Of_Plenty_Per_Capita  = ownership$Bay_Of_Plenty / populations$Bay_Of_Plenty,
                Canterbury_Per_Capita = ownership$Canterbury / populations$Canterbury,
                Gisbourne_Per_Capita = ownership$Gisbourne / populations$Gisbourne,
                Hawkes_Bay_Per_Capita = ownership$Hawks_Bay / populations$Hawks_Bay,
                Taranaki_Per_Capita = ownership$Taranaki / populations$Taranaki,
                Tasman_Per_Capita  = ownership$Tasman / populations$Tasman,
                Waikato_Per_Capita = ownership$Waikato / populations$Waikato,
                Wellington_Per_Capita = ownership$Wellington / populations$Wellington,
                West_Coast_Per_Capita = ownership$West_Coast / populations$West_Coast,
                Manawatu_Whanganui_Per_Capita = ownership$Manawatu_Whanganui / populations$Manawatu_Whanganui,
                Marlborough_Per_Capita  = ownership$Marlborough / populations$Marlborough,
                Nelson_Per_Capita = ownership$Nelson / populations$Nelson,
                Otago_Per_Capita = ownership$Otago / populations$Otago,
                Southland_Per_Capita = ownership$Southland / populations$Southland,
                Northland_Per_Capita = ownership$Northland / populations$Northland
                
                  )
ownership_with_pc
```

And finally take out just the columns that we want 

```r
#And finally make a new dataframe by extracting just the per capita informatino from the previous dataframe 
homes_pc = ownership_with_pc[,c(17:32)]
homes_pc
```

And then add some row names back in to make this a more meaningful dataframe

```r
row.names(homes_pc) <- c("House Owned or Partially Owned","House Not Owned")
```
