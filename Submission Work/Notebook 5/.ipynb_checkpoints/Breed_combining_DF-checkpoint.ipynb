### Sort breeds based on appearance in the Figures NZ data, used Breeds_Difference.csv as reference

```r
library(dplyr)

# create a vector with letters in the desired order, in same order as Figures Breeds
akc_breeds <- c(7, 43, 14, 55, 35, 136, 33, 38, 162, 36, 6, 3, 1, 8, 18, 2, 20, 16, 28, 27, 61, 122, 75, 82, 44)

ordered_breeds <- filtered_breeds %>%
  slice(match(akc_breeds, Rank))

ordered_breeds
```

### Trim whitespace from scraped dog breeds. Looks like Beagle hasn't been trimmed, investigated the CSV and Beagle input format was the same as all other breeds 

```r
ordered_breeds$Breed <- trimws(ordered_breeds$Breeds, which = c("both"))

sorted_akc_breeds= subset(ordered_breeds, select = -c(Breeds) )
write.csv(sorted_akc_breeds, "sorted_akc_breeds.csv")
sorted_akc_breeds
```
