### Working to get the list of dog breeds

```r
dog_ranking_url <- "https://www.akc.org/expert-advice/dog-breeds/2020-popular-breeds-2019/"
```

```r
ranking_page <- read_html(dog_ranking_url)
```

```r
ranking_page %>%
  html_nodes("#main-content > div.article-body > div.content-body > div > div > table > tbody > tr:nth-child(1) > td:nth-child(1)")
```

```r
doggo_df <- tibble(Rank = 1:193, 
                  Link = ranking_page %>%
  html_nodes(".in-cell-link") %>%
  html_attrs() %>%
  map_chr("href"))

doggo_df
```

```r
kennel_breed <- glue("{doggo_df$Link[1]}")

kennel_breed
```

```r
kennel_breed %>%
  extract_breed_from_url
```

```r
get_all_breed <- function(link){
    
    breed_names<- glue("{link}") %>%
          read_html() %>%
          html_node("body") %>%
          html_nodes("#page-title > h1") %>%
          html_text()
    
    breeds <- gsub("\n","",breed_names)
    
    return(breeds)
}
```

```r
extract_height_f_url <- function(url){
  text_part <- url %>%
  read_html() %>%
  rvest::html_node("body") %>%
  rvest::html_nodes("#panel-overview > div > div > div.panel-flex__aside > ul > li:nth-child(3) > span.attribute-list__description.attribute-list__text") %>%
  rvest::html_text()
    
  return(text_part)
}
```

```r
extract_weight_f_url <- function(url){
  text_part <- url %>%
  read_html() %>%
  rvest::html_node("body") %>%
  rvest::html_nodes("#panel-overview > div > div > div.panel-flex__aside > ul > li:nth-child(4) > span.attribute-list__description.attribute-list__text") %>%
  rvest::html_text()
    
  return(text_part)
}
```

```r
extract_item_f_url <- function(url, css_selector){
  text_part <- url %>%
  read_html() %>%
  rvest::html_node("body") %>%
  rvest::html_nodes(css_selector) %>%
  rvest::html_text()
    
  return(text_part)
}
```
